> # **This function prints the game board to the console**

```python
def output(game):
    for i in range(3):
        for j in range(3):
            print(game[i][j],end="|")
        print()
```

> # **this function updates the game board with the player's symbol in the appropriate location**

```python
def update(game , place,symbol):
    move={1:(0,0),2:(0,1),3:(0,2),4:(1,0),5:(1,1),6:(1,2),7:(2,0),8:(2,1),9:(2,2)}
    i,j=move[place]
    if game[i][j]==' ':
        game[i][j]=symbol
        return True
    return False
```

> # **This function checks for a win**

```python
def win(game,place,symbol):
    move={1:(0,0),2:(0,1),3:(0,2),4:(1,0),5:(1,1),6:(1,2),7:(2,0),8:(2,1),9:(2,2)}
    i,j=move[place]
    test=[symbol for i in range(3)]
    row=game[i]
    col=[game[k][j]  for k in range(3)]
    diag1=[]
    diag2=[]
    if(i==j):
        diag1=[game[i][i] for i in range(3)]
    if(i+j==2):
        diag2=[game[i][2-i] for  i in range(3)]
    if(row==test):
        return True
    elif (col==test):
        return True
    elif (diag2==test or diag1==test):
        return True
    return False
```

> # **This function to check for a tie**

```python
def Tie(game):
    for i in range(3):
        if(' '  in game[i]):
            return False
    return True
```

> # **Main game program**

```python
game=[[' ']*3 for i  in  range(3)]
print("start the game with the first player 'X' ")
move={1:(0,1),2:(0,2),3:(0,3),4:(1,0),5:(1,1),6:(1,2),7:(2,0),8:(2,1),9:(2,2)}
i=0;
output(game)
while(1):
    if (i%2==0):
        symbol='X'
    else:
        symbol='O'
    place=int(input(f"player {symbol}:"))
    if not(update(game , place,symbol)):
        print('the place you have entered is full,try again:')
    else:
        output(game)
        if win(game,place,symbol):
            print(f"congratulation ,the player  {symbol} wins the game")
            break
        elif Tie(game):
            print("the game has over")
            break
        i+=1
```
